```python
import pandas as pd 
dataset = pd.read_csv('./Dataset DAMI.csv')
```

```python
dataset.sample(10)
```

```python
# Feature vector dan target variable
X = dataset.drop(['Label'], axis=1)

y = dataset['Label']
```

```python
# Import fungsi train_test_split 
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier 
dt = DecisionTreeClassifier()
dt = dt.fit(X_train,y_train)
```

```python
from sklearn import metrics
y_pred = dt.predict(X_train)
print("Accuracy:",metrics.accuracy_score(y_train, y_pred))
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
y_pred = dt.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
dt = DecisionTreeClassifier(criterion="entropy", max_depth=3)
dt = dt.fit(X_train,y_train)
y_pred = dt.predict(X_train)

print("Train accuracy score dengan max_depth=3 {0:0.4f}".format(accuracy_score(y_train, y_pred)))
```

```python
dt = DecisionTreeClassifier(criterion="entropy", max_depth=3)
dt = dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)

print("Test accuracy score dengan max_depth=3 {0:0.4f}".format(accuracy_score(y_test, y_pred)))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
#make a prediction
dt = DecisionTreeClassifier(criterion="entropy", max_depth=5)
dt = dt.fit(X_train,y_train)
y_pred = dt.predict(X_train)
```

```python
print("Train accuracy score dengan max_depth=5 {0:0.4f}".format(accuracy_score(y_train, y_pred)))
```

```python
print("Test accuracy score dengan max_depth=5 {0:0.4f}".format(accuracy_score(y_test, y_pred)))
```
